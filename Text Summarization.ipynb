```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
```

```python
import bs4 as bs
import urllib.request
```

```python
sourse = urllib.request.urlopen('https://en.wikipedia.org/wiki/Global_warming').read()
```

```python
sourse
```

```python
soup = bs.BeautifulSoup(sourse, 'lxml')
```

```python
soup
```

```python
text = ""

for para in soup.find_all('p'):
    text = text + para.text
```

```python
text
```

```python
text1 = []

for para in soup.find_all('p'):
    text1.append(para.text)
```

```python
text1
```

```python
import re
```

```python
text = re.sub(r'\[[0-9]*\]', ' ', text)
text = re.sub(r'\s+', ' ', text)
```

```python
ctext = text.lower()
ctext = re.sub(r'\W', ' ', ctext)
ctext = re.sub(r'\d', ' ', ctext)
ctext = re.sub(r'\s+', ' ', ctext)
ctext = re.sub(r'\s+[a-z]\s+', ' ', ctext)
```

```python
sentences = nltk.sent_tokenize(text)
```

```python
sentences[0]
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words = nltk.corpus.stopwords.words('english')
```

```python
stop_words
```

```python
word_count = {}

for word in nltk.word_tokenize(ctext):
    if word not in stop_words:
        if word not in word_count.keys():
            word_count[word] = 1
        else:
            word_count[word] = word_count[word] + 1
```

```python
word_count
```

```python
for key in word_count.keys():
    word_count[key] = word_count[key] / max(word_count.values())
```

```python
word_count
```

```python
final = {}

for i in range(len(sentences)):
    score = 0
    
    for word in nltk.word_tokenize(str(sentences[i]).lower()):
        if len((sentences[i]).split(' ')) < 30:
            for key in word_count.keys():
                if word == key:
                    score = score + word_count[key]
                else:
                    score = score
                
    final[sentences[i]] = score
```

```python
import heapq

best_sent = heapq.nlargest(10, final, key = final.get)
```
